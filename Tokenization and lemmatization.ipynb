# Расчет TF-IDF 

Рассчитать TF-IDF используя язык программирования Python с помо-щью класса Counter из стандартной библиотеки collections

```
text_tf_ifd = '''
People divorce Donald Trump the owner of the golf course from his politics,” said James Griffin, a member of the Trump club interviewed by the Irish Times. At Doonbeg, Trump is likely to find something that escaped him in London: a warm welcome. Trump’s club employs more than 200 people, making it one of the largest employers in a rural area of County Clare.The Trump Organization cited climate change in its application for the permit, according to a Politico report from 2016, saying that sea-level rise and more-powerful storms had worsened the threat of erosion. Trump the politician, of course, has questioned idea that climate change is a threat at all — defying the overwhelming scientific consensus and his own golf course’s assessment of its fu-ture.
'''
```

```
import collections
from collections import Counter
import math
```

```
def compute_tfidf(corpus):
  
  # Рассчитаем TF 
  def compute_tf(text):
    tf_text = Counter(text)
    for i in tf_text:
      tf_text[i] = tf_text[i]/float(len(text))
    return tf_text

  # Рассчитаем IDF
  def compute_idf(word, corpus):
    return math.log10(len(corpus)/sum([1.0 for i in corpus if word in i]))
  
  # Список из словарей: слово - значение TF-IDF
  documents_list = []

  # corpus - список из списков слов, text - список слов 1 предложения
  for text in corpus:
    tf_idf_dictionary = {}
    computed_tf = compute_tf(text)

    for word in computed_tf:
      tf_idf_dictionary[word] = computed_tf[word] * compute_idf(word, corpus)
    
    documents_list.append(tf_idf_dictionary)

  return documents_list
```

```
# Весь текст в нижний регистр
text_tf_ifd = text_tf_ifd.lower()
# Разобъем текст на предложения 
sentences = nltk.sent_tokenize(text_tf_ifd) 
# Разобьем предложения на слова
corpus = []
tokenizer = nltk.RegexpTokenizer(r"\w+")
for i in sentences:
  list_tmp = tokenizer.tokenize(i)
  corpus.append(list_tmp)
```

```
res_dic = compute_tfidf(corpus)
res_dic
```
