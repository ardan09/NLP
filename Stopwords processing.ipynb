```
text = text.lower()
text
```

```
#import string
import re
```

```
#string.punctuation
```

```
words = re.split(r'\W+', text)
words
```

```
!pip install pymorphy2
```

```
import pymorphy2
pymorph = pymorphy2.MorphAnalyzer()
```

```
new_words = []
for word in words:
   word = pymorph.parse(word)[0].normal_form 
   new_words.append(word)
```

```
new_words = [i for i in new_words if i != '']
new_words
```

#### 1ый способ выделения стоп-слов

```
stop_words_1 = [i for i in new_words if i in nltk_russian_stopwords]
stop_words_1
```

```
 len(stop_words_1)
```

```
from stop_words import get_stop_words

stop_words = get_stop_words('russian')
```

```
stop_words_2 = [i for i in new_words if i in stop_words]
stop_words_2
```
